3. As motos que possuiram mais donos são as motos mais velhas na média?

```python
# Função que fará o calculo da idade da motocicleta
def calculate_age(year):
    # Recupera a data de hoje
    today = date.today()
    
    # Como estamos calculando somente a idade através dos anos, podemos
    # simplismente subtrair o ano da data atual pelo ano da moto
    return today.year - year
```

```python
# Cria a coluna 'age', que conterá a idade da motocicleta
df['age'] = df['year'].apply(lambda year: calculate_age(year))
df.head()
```

```python
df_grouped = df[['owner', 'age']].groupby('owner').mean().sort_values('age', ascending=False).reset_index()

df_grouped
```

Sim, As motos que possuiram mais donos são as motos mais velhas na média

5. O CEO lhe entregou um novo dataset chamado companies.csv, onde estão todas as fabricantes de motocicletas. Adicione uma coluna no DataFrame com o nome de company. Essa coluna deve possuir o nome do fabricante de cada moto do DataFrame.

```python
# Lendo o dataset de fabricantes
df_companies = pd.read_csv('csv//companies.csv', index_col=0)
#df = pd.read_csv('csv//bike.csv')

df_companies.head()
```

```python
def create_sorted_companies_tuple(df_companies):
    companies = df_companies['company_name'].to_list() # Passou nomes do df para lista
    #print(companies)
    #print()
    #print() 

    # Utilizando List Comprehensions

    companies_aux = [ company.split(' ') for company in companies ] # Cada nome passa a ter sua propria lista (sublista geral) de palavras
    #print(companies_aux)
    #print()
    #print()

    companies_aux.sort(key=len, reverse=True) # executou uma ordenacao na lista pelo tamanho de palavras (sublistas)
    #print(companies_aux)
    #print()
    #print()

    sorted_companies = [ ' '.join(company) for company in companies_aux ] # Passa a ser uma unica lista (sem sublista) sem espaços e ordenada das maiores strings para as menores
    #print(sorted_companies)

    return tuple(sorted_companies)


```

```python

def create_company_name(companies_tuple, data): # Recebe a tupla ordenada das maiores strings para as menores
   
    for company in companies_tuple: #pra cada elemento na tupla faca
        if company in data['name']: # se a company estiver contida em data['name']
            return company # retorna a company
```

```python


# Método novo
companies_tuple = create_sorted_companies_tuple(df_companies) # chama a funcao que tras a tupla de company ordenada
#print(companies_tuple)


# Criando a coluna `company` no DataFrame
df['company'] = df.apply(lambda line: create_company_name(companies_tuple, line), axis=1)
```
